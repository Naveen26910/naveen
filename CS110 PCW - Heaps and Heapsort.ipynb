Write a Python implementation of the MAX-HEAPIFY operation using the pseudocode above, and your newly written functions, `left` and `right`.

```python
def heapify(arr, n, i): #online sources used for help
    largest = i  # largest value as root of the binary tree
    l = 2 * i + 1     # index starting left = 2*i + 1
    r = 2 * i + 2     # if swaps need to be made from right node 

    if l < n and arr[largest] < arr[l]: #if child of rote node at the left side of the tree is greater than root
        largest = l
 
    if r < n and arr[largest] < arr[r]: #if right side node is greater than parent node/root
        largest = r

    if largest != i: #root is changed if node has greater value
        arr[i], arr[largest] = arr[largest], arr[i] 
        heapify(arr, n, largest)


def heapSort(arr):
    n = len(arr)
 
   
    for i in range(n//2 - 1, -1, -1): #building max heap
        heapify(arr, n, i)
 
    for i in range(n-1, 0, -1): #arranging elements one by onw
        arr[i], arr[0] = arr[0], arr[i]  # swapping elements if not in order of max heap binary tree
        heapify(arr, i, 0) #using heapify function for array indexed at 0 
        
arr = [39, 85, 85, 16, 49, 7, 49, 92, 76, 15, 21, 30, 29, 31, 28] #testing code using unsorted array
heapSort(arr)
n = len(arr)
print("maximum heap is")
for i in range(n):
    print("%d" % arr[i]),
```
